```python
import cartopy.crs as ccrs
import cartopy.feature
import matplotlib.pyplot as plt
import calendar
import polars as pl
import cmocean

import parcels
```

```python
fig, axes = plt.subplots(4, 3, figsize=(10, 7.5), constrained_layout=True, subplot_kw={'projection': ccrs.PlateCarree()})

cmap = cmocean.tools.crop_by_percent(cmocean.cm.speed, 20, which='min', N=None)

for i, ax in enumerate(axes.flatten()):
    df = parcels.read_particlefile(f"Simulations/Simulation_Basin_2024-{i+1:02d}_Lagrangian.parquet")

    to_plot = (
        df.group_by('particle_id')
        .agg([
            pl.col('lon').first().alias('lon'),
            pl.col('lat').first().alias('lat'),
            pl.col('biomass_SF3').last().alias('biomass')
        ])
        .sort('particle_id')
    )
    cscat = ax.scatter(
        to_plot['lon'], to_plot['lat'], c = to_plot['biomass'] * 100 ,
        cmap=cmap, s=4, linewidth=0, transform=ccrs.PlateCarree(), zorder=2,
        vmin=0.0, vmax=500,
    )

    ax.add_feature(cartopy.feature.COASTLINE.with_scale('10m'), zorder=3, color='black')
    ax.add_feature(cartopy.feature.LAND.with_scale('10m'), zorder=2, color='linen')
    ax.set_extent([-75, -11, -3, 20])
    ax.set_title(calendar.month_name[i+1], fontsize=11)

cbar = fig.colorbar(cscat, ax=axes, orientation='horizontal', pad=0.04, aspect=50, label='Lagrangian biomass change [%]')

plt.savefig("Figures/Figure8.png", bbox_inches="tight", dpi=300)
plt.show()
```
